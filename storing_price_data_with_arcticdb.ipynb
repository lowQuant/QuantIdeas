```python
import time
import numpy as np
import pandas as pd
from datetime import datetime
from arcticdb import Arctic, QueryBuilder

import yfinance as yf
```

### Creating the arcticdb object

```python
ac = Arctic('lmdb://MyDB?map_size=2GB')
```

### Structure of arcticdb

![Alt text](./adb_schema.png)


### Create a library

```python
library = ac.get_library('spx500', create_if_missing=True)
```

```python
# Show all libraries
ac.list_libraries()
```

### Write into arcticdb

```python
df = yf.download("NFLX")
df.tail()
```

```python
library.write(symbol="NFLX",data=df)
```

#### Read from arcticdb

```python
query = library.read(symbol="NFLX") # can take an additional parameter "as_of" for point in time data
query.data
```

### Download & write the S&P500 symbols

```python
spx500 = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies",index_col=0)[0]
symbols = spx500.index.to_list()
```

```python
data = yf.download(symbols, group_by="Ticker", start="2022-12-31")
```

```python
data.head()
```

```python
df = data.stack(level=0).rename_axis(['Date', 'Ticker']).reset_index(level=1)
df.head()
```

```python
for symbol in df["Ticker"].unique():
    library.write(symbol=symbol, data= df[df["Ticker"]==symbol])
```

```python
library.read(symbol="AMZN").data
```

### Bulk Write - not really faster...

```python
from arcticdb.version_store.library import WritePayload
```

```python
dataframes = []

for symbol in df["Ticker"].unique():
    dataframes.append(WritePayload(symbol,df[df["Ticker"]==symbol]))
        
library.write_batch(dataframes)

```

### Query Data

```python
q = QueryBuilder()
# look for shares with a trading volume above 1m, where price is <= 10 USD
q = q[(q["Volume"]>1e6) & (q["Adj Close"] <= 10000)]
filtered = library.read("AAPL", query_builder=q).data
filtered
```
